## 1. Write a function that returns the greater of two numbers

```python
def greater(a, b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        print("The two numbers are equal")
        return a
```

```python
# This will test your function 
test_greater(greater)
```

```python
class TestGreaterFunction(unittest.TestCase):

    def test_greater_first_is_bigger(self):
        self.assertEqual(greater(10, 5), 10)

    def test_greater_second_is_bigger(self):
        self.assertEqual(greater(3, 8), 8)

    def test_greater_equal_numbers(self):
        self.assertEqual(greater(7, 7), 7)

# Run tests
if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
```

## 2. Now write a function that returns the largest element on a list

```python
def greatest(lst):
    largest = lst[0]
    i = 0
    while i < len(lst) :
        if lst[i] > largest:
            largest = lst[i]
        i +=1

    return largest
```

```python
# This will test your function 
test_greatest(greatest)
```

```python
greatest([500, 150, 100,1000])
```

## 3. Write a function that sums all the elements on a list

```python
def sum_all(lst):
    summ = 0
    i = 0
    while i < len(lst) :
        summ +=lst[i]
        i +=1

    return summ
```

## 4. Write another function that multiplies all the elements on a list

```python
def mult_all(lst):
    multiplies = 1 
    i = 0
    while i < len(lst) :
        multiplies *= lst[i]
        i += 1

    return multiplies
```

```python
def oper_all(arr, oper):
    if oper == "+" :
        return sum_all(arr)
    elif oper == "*" :
        return mult_all(arr)
    else :
        raise ValueError("Operator must be '+' or '*'")
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    i = 0
    factorial = 1
    while i < n :
        i += 1
        factorial *= i
    return factorial
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

```python
factorial(10)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
    list_unique = []
    for i in lst_un :
        if i not in list_unique :
            list_unique.append(i)
        else :
            pass
    return list_unique
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
    freq = {} 

    for item in arr:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    max_freq = 0
    mode_value = None

    for item, f in freq.items():
        if f > max_freq:
            max_freq = f
            mode_value = item

    return mode_value
```

```python
# This will test your function 
test_mode(mode_counter)
```

```python
mode_counter([1000, 500, 150, 100,1000, 150, 100,1000, 100,1000])
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):
    # mean
    total = 0
    for x in list_sd:
        total += x
    mean = total / len(list_sd)
    
    # variance
    total_sq = 0
    for x in list_sd:
        total_sq += (x - mean) ** 2
    variance = total_sq / (len(list_sd) - 1)
    
    return variance**0.5
```

```python
# This will test your function 
test_stdev(st_dev)
```

```python
st_dev([1000, 500, 150, 100,1000, 150, 100,1000, 100,1000])
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    found = []

    for char in string.lower():
        if char in alphabet and char not in found:
            found.append(char)

    return len(found) == 26
```

```python
# This will test your function 
test_pangram(pangram)
```

```python
pangram("The quick brown fox jumps over the lazy dog")
```

```python
pangram("Hello world")
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
    # Step 1: découper la chaîne manuellement
    words = []
    current = ""

    for char in string:
        if char == ",":
            words.append(current)
            current = ""
        else:
            current += char

    # Ajouter le dernier mot
    words.append(current)

    # Step 2: trier
    sorted_words = sorted(words)

    # Step 3: reconstruire la chaîne sans join
    result = ""
    for i in range(len(sorted_words)):
        result += sorted_words[i]
        if i < len(sorted_words) - 1:
            result += ","

    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

```python
sort_alpha("pear,apple,banana")
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    if len(password) < 8:
        return False

    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False

    specials = "!@#$%^&*()-_=+[]{};:'\",.<>/?\\|`~"

    for char in password:
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_digit = True
        elif char in specials:
            has_special = True

    return has_lower and has_upper and has_digit and has_special
```

```python
# This will test your function 
test_pass(check_pass)
```

```python
check_pass("Hello123!")
```

```python
check_pass("weakpass")
```
